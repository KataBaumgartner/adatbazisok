```sql
SELECT  CASE WHEN LEIRAS IS NULL THEN 'Null'
                  ELSE 'Leírás' 
                  END AS 'Van leírás vagy null?',
        COUNT(*) AS 'Darabszám'
FROM Termek
GROUP BY CASE WHEN LEIRAS IS NULL THEN 'Null' ELSE 'Leírás' END
```

Kiugró értékek, átlagolás

```sql
SELECT MIN(LISTAAR) AS 'Legalacsonyabb ár',
       MAX(LISTAAR) AS 'Legmagasabb ár', 
       AVG(LISTAAR) AS 'Átlag ár', 
       STDEV(LISTAAR) AS 'Árak szórása', 
       VAR(LISTAAR) AS 'Árak varianciája'
FROM Termek

```

```sql
SELECT LEN(MEGNEVEZES) AS 'Megnevezés hossza',
       count(*) AS 'Darab'
FROM Termek
GROUP BY LEN(MEGNEVEZES)
ORDER BY 1
```

```sql
SELECT MEGYS AS 'Mértékegység', 
        LEN(MEGYS) AS 'String hossza'
FROM Termek
GROUP BY MEGYS
ORDER BY LEN(MEGYS) DESC
```
